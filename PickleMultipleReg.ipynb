```python
import pandas as pd
import pickle

# Load the Pickle
model = pickle.load(open('multiple_reg.pkl','rb'))      # r = read b = byte stream
```

```python
# Final Prediction
california = int(input('Is startup is in California ? Enter 1 if Yes else 0 : '))
florida = int(input('Is startup is in Florida ? Enter 1 if Yes else 0 : '))
newYork = int(input('Is startup is in NewYork ? Enter 1 if Yes else 0 : '))
rdSpend = int(input('Enter R&D Spend : '))
adminSpend = int(input('Enter Administration Spend : '))
marketingSpend = int(input('Enter Marketing Spend : '))

user_input = {'california':california , 'florida':florida , 'newYork':newYork ,
              'rdSpend':rdSpend  , 'adminSpend':adminSpend , 'marketingSpend':marketingSpend}

user_data = pd.DataFrame([user_input],index=[0])
user_data = user_data.values
print(user_data)
```

```python
predict_profit = model.predict(user_data)
print('Predicted Profit : ',int(predict_profit[0]))
```
